```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

```python
plt.figure(figsize=(12, 5))

# Left panel for sin(x)
plt.subplot(1, 2, 1)
plt.plot(x, sin_function(x), label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin(x) vs. x')
plt.legend()
plt.grid(True)

# Right panel for cos(x)
plt.subplot(1, 2, 2)
plt.plot(x, cos_function(x), label='cos(x)', color='orange')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('cos(x) vs. x')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
