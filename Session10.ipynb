```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 1000 random numbers from the normal distribution with mean 0 and standard deviation 1
random_data = np.random.normal(0, 1, 1000)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(random_data, bins=100, color='green', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers (Normal Distribution)')
plt.grid(True)
plt.show()
```
