```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_data = np.random.uniform(0, 1, 1000)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(random_data, bins=100, color='blue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers (Uniform Distribution)')
plt.grid(True)
plt.show()
```
